```python
import requests
from lxml import html
from urllib import parse
import smtplib
import time
import pandas as pd
import numpy as np
import re
```

```python
avito = 'https://www.avito.ru'
no_years = 0
old_money = 0
new_money = 0
i = 0
```

```python
def find_first_year(s):
    l = re.findall('\D{1}(\d{4})\D{1}', ' ' + s + ' ')
    l = [i for i in l if int(i) > 1000 & int(i) < 2018]
    return l
```

```python
for s in ['&pmax=1000', '&pmin=1001']:
    monety = '/moskva/kollektsionirovanie/monety?view=list' + s
    while True:
        print(monety)
        r = requests.get(avito + monety)
        content = html.fromstring(r.content)
        items = content.xpath('//a[@class="description-title-link"]')
        descriptions = [i.xpath('descendant::text()')[0] for i in items]
        years = [find_first_year(s) for s in descriptions]
        for year in years:
            if len(year) == 0:
                no_years += 1
            else:
                y = int(year[0]) 
                if y < 2000:
                    old_money += 1
                else:
                    new_money += 1
        print('\t {} {} {}'.format(no_years, old_money, new_money))
        page = content.xpath('//a[@class="pagination-page js-pagination-next"]')
        if len(page) == 0:
            break
        monety = page[0].xpath('.//@href')[0]
        i += 1
```
